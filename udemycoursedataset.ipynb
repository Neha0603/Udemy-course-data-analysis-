```python
#Find out the highest price of course?
course['price'].max()

#Find out the lowest price of course?
course['price'].min()
```

```python
# which course has the highest price?
course[course['price']==course['price'].max()].head(2)
```

```python
# Calculate which course has the highest reviews?
course[course['num_reviews']==course['num_reviews'].max()]
```
